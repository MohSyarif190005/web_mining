#### Cleaning Text
1. Fungsi cleaningText, menghapus mention, menghapus hashtag, menghapus RT, menghapus link, dan menghapus nomor, mengganti baris baru dengan spasi, menghapus semua tanda baca, dan menghapus berbagai macam karakter atau emoji teks dari kiri ke kanan. <br>

```python
def cleaningText(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) # remove mentions
    text = re.sub(r'#[A-Za-z0-9]+', '', text) # remove hashtag
    text = re.sub(r'RT[\s]', '', text) # remove RT
    text = re.sub(r"http\S+", '', text) # remove link
    text = re.sub(r'[0-9]+', '', text) # remove numbers

    text = text.replace('\n', ' ') # replace new line into space
    text = text.translate(str.maketrans('', '', string.punctuation)) # remove all punctuations
    text = text.strip(' ') # remove characters space from both left and right text
    return text
```

#### Case Folding Text

2. Fungsi caseFoldingText, mengubah semua huruf menjadi huruf kecil (lowercase) semua. Sementara itu, karakter lain yang bukan termasuk huruf dan angka, seperti tanda baca dan spasi dianggap sebagai delimiter. Delimiter ini bisa juga dihapus atau diabaikan dengan menggunakan perintah yang ada di Python.<br>

```python
def casefoldingText(text): # Converting all the characters in a text into lower case
    text = text.lower() 
    return text
```

#### Tokenizing Text
3. Fungsi tokenizingText, pemisahan string pada teks menjadi beberapa daftar token sehingga dapat digunakan untuk menganalisa teks <br>

```python
def tokenizingText(text): # Tokenizing or splitting a string, text into a list of tokens
    text = word_tokenize(text) 
    return text
```

#### Filtering Text
4. Fungsi filteringText, tahap mengambil kata-kata penting dari hasil token dengan menggunakan algoritma stoplist (membuang kata kurang penting) atau wordlist (menyimpan kata penting).


```python
def filteringText(text): # Remove stopwors in a text
    listStopwords = set(stopwords.words('indonesian'))
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    text = filtered 
    return text
```

#### Stemming Text
5. Fungsi stemmingText, proses menghilangkan infleksi kata ke bentuk dasarnya, namun bentuk dasar tersebut tidak berarti sama dengan akar kata (root word).

```python
def stemmingText(text): # Reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    text = [stemmer.stem(word) for word in text]
    return text
```

#### to Sentence
6. Fungsi toSentence, Ubah daftar kata menjadi kalimat


```python
def toSentence(list_words): # Convert list of words into sentence
    sentence = ' '.join(word for word in list_words)
    return sentence
```

Memanggil Semua fungsi Preprocessing untuk melakukan preprocessing data

```python
# Preprocessing tweets data

tweets['text_clean'] = tweets['tweet'].apply(cleaningText)
tweets['text_clean'] = tweets['text_clean'].apply(casefoldingText)
tweets.drop(['tweet'], axis = 1, inplace = True)

tweets['text_preprocessed'] = tweets['text_clean'].apply(tokenizingText)
tweets['text_preprocessed'] = tweets['text_preprocessed'].apply(filteringText)
tweets['text_preprocessed'] = tweets['text_preprocessed'].apply(stemmingText)

# drop duplicates/spams tweets
tweets.drop_duplicates(subset = 'text_clean', inplace = True)
```

## Term Frequency (TF)
Term Frequency — Inverse Document Frequency atau TF — IDF adalah suatu metode algoritma yang berguna untuk menghitung bobot setiap kata yang umum digunakan. Metode ini juga terkenal efisien, mudah dan memiliki hasil yang akurat. Metode ini akan menghitung nilai Term Frequency (TF) dan Inverse Document Frequency (IDF) pada setiap token (kata) di setiap dokumen dalam korpus. Secara sederhana, metode TF-IDF digunakan untuk mengetahui berapa sering suatu kata muncul di dalam dokumen.

```python
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer, CountVectorizer
# get bag of words features in sparse format
vectorizer = CountVectorizer(min_df=1)
X = vectorizer.fit_transform(tweets_clean_label['text_preprocessed'])
z = vectorizer.get_feature_names()
count=pd.DataFrame(X.toarray(),columns=z)
count
```

### Metode K-NN
Algoritma KNN merupakan algoritma klasifikasi yang bekerja dengan mengambil sejumlah K data terdekat (tetangganya) sebagai acuan untuk menentukan kelas dari data baru. Algoritma ini mengklasifikasikan data berdasarkan similarity atau kemiripan atau kedekatannya terhadap data lainnya.<br>
Secara umum, cara kerja algoritma KNN adalah sebagai berikut.
1. Tentukan jumlah tetangga (K) yang akan digunakan untuk pertimbangan penentuan kelas.<br>
2. Hitung jarak dari data baru ke masing-masing data point di dataset. <br>
3. Ambil sejumlah K data dengan jarak terdekat, kemudian tentukan kelas dari data baru tersebut. <br>


```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

## Ensemble Stacking

### Pengertian Ensemble Stacking
Ensemble Stacking adalah  metode yang menggunakan beberapa algoritma atau model pembelajaran untuk menghasilkan satu model prediksi yang optimal. Model yang dihasilkan memiliki kinerja yang lebih baik dibandingkan dengan basis peserta didik yang diambil sendiri. Aplikasi lain dari pembelajaran ensemble juga termasuk memilih fitur penting, fusi data, dll. Teknik ensemble dapat diklasifikasikan menjadi Bagging, Boosting, and Stacking. <br>

<img src='https://editor.analyticsvidhya.com/uploads/27473ensemble.png'>

### Algoritma Ensemble Stacking
Bagging dan boosting menggunakan pelajar lemah yang homogen untuk ensamble, Stacking sering menganggap pelajar lemah yang heterogen, mempelajarinya secara paralel, dan menggabungkannya dengan melatih meta-pelajar untuk menghasilkan prediksi berdasarkan prediksi pelajar lemah yang berbeda. Pelajar meta memasukkan prediksi sebagai fitur dan target menjadi nilai kebenaran dasar dalam data D (Gbr 2.), ia mencoba mempelajari cara terbaik untuk menggabungkan prediksi input untuk membuat prediksi output yang lebih baik.

Dalam ansambel rata-rata misalnya. Random Forest, model menggabungkan prediksi dari beberapa model terlatih. Keterbatasan pendekatan ini adalah bahwa setiap model memberikan kontribusi jumlah yang sama untuk prediksi ensemble, terlepas dari seberapa baik model dilakukan. Pendekatan alternatif adalah ensemble rata-rata tertimbang, yang menimbang kontribusi setiap anggota ensemble dengan kepercayaan atas kontribusi mereka dalam memberikan prediksi terbaik. Ansambel rata-rata tertimbang memberikan peningkatan atas ansambel rata-rata model.

Generalisasi lebih lanjut dari pendekatan ini adalah mengganti jumlah tertimbang linier dengan Regresi Linier (masalah regresi) atau Regresi Logistik (masalah klasifikasi) untuk menggabungkan prediksi sub-model dengan algoritma pembelajaran apa pun. Pendekatan ini disebut Stacking.

### Pengimplementasian
Dalam Ensemble Stacking ini, dalam eksperimen mencoba berbagai macam klasifikasi. Diantaranya Gaussian Naive Bayes, Random Forest Classifier yang sudah dicari nilai parameter yang terbaik menggunakan GridSearhCV, SVM, KNN dengan nilai n sebesar 12, Bagging Classifier dengan nilai terbaik menggunakan GridSearchCV. Dengan nilai Final estimatornya ialah menggunakan Gaussian Naive Bayes

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'base_estimator__max_depth' : [1, 2, 3, 4, 5],
    'max_samples' : [0.05, 0.1, 0.2, 0.5]
}

estimators = [
    ('rf', RandomForestClassifier(max_depth=3, max_features=3, min_samples_leaf=20,
                       n_estimators=10)),
    ('GNB', GaussianNB()),
    ('SVM', svm.SVC()),
    ('knn', KNeighborsClassifier(n_neighbors=14))
]
clf = StackingClassifier(
    estimators=estimators, final_estimator=svm.SVC()
)

clf.fit(X_train, y_train).score(X_test, y_test)
```
