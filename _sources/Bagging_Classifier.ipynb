# Bagging Classifier (Ensemble Learning), Grid Search dan Random Forest

## Pengertian Bagging Classifier
Pengklasifikasi Bagging adalah meta-estimator ensemble yang cocok dengan pengklasifikasi dasar masing-masing pada subset acak dari dataset asli dan kemudian menggabungkan prediksi individu mereka (baik dengan pemungutan suara atau rata-rata) untuk membentuk prediksi akhir. Meta-estimator seperti itu biasanya dapat digunakan sebagai cara untuk mengurangi varians dari estimator kotak hitam (misalnya, pohon keputusan), dengan memasukkan pengacakan ke dalam prosedur konstruksinya dan kemudian membuat ansambel darinya.
Setiap pengklasifikasi dasar dilatih secara paralel dengan set pelatihan yang dihasilkan dengan menggambar secara acak, dengan penggantian, N contoh (atau data) dari set data pelatihan asli – di mana N adalah ukuran set pelatihan asli. Pelatihan set untuk masing-masing pengklasifikasi dasar adalah independen satu sama lain. Banyak dari data asli dapat diulang dalam set pelatihan yang dihasilkan sementara yang lain mungkin ditinggalkan. <br>

Bagging mengurangi overfitting (varians) dengan rata-rata atau voting, namun, hal ini menyebabkan peningkatan bias, yang dikompensasi oleh pengurangan varians sekalipun. <br>

Bagaimana Bagging bekerja pada dataset pelatihan?
Cara kerja bagging pada dataset pelatihan imajiner ditunjukkan di bawah ini. Karena Bagging membuat sampel ulang set data pelatihan asli dengan penggantian, beberapa instance (atau data) mungkin ada beberapa kali sementara yang lain tidak disertakan. <br>

## Algoritma Bagging Classifier
Algoritma ini mencakup beberapa karya dari literatur. Ketika subset acak dari dataset diambil sebagai subset acak dari sampel, maka algoritma ini dikenal sebagai Paste. Jika sampel diambil dengan penggantian, maka metode tersebut dikenal sebagai Bagging. Ketika subset acak dari dataset digambar sebagai subset acak dari fitur, maka metode ini dikenal sebagai Random Subspaces. Akhirnya, ketika penduga dasar dibangun di atas himpunan bagian dari sampel dan fitur, maka metode ini dikenal sebagai Patch Acak.

## Pengertian Grid Search
Grid Seacrh, Metode ini paling mudah untuk diterapkan dan dipahami, tetapi sayangnya tidak efisien ketika jumlah parameternya besar dan tidak terlalu dibatasi di bawah H0. Biarkan Ω* menjadi ruang parameter gangguan = (ν1, 2, …νm) di mana kita memaksimalkan nilai-p. Cara sederhana untuk mengatur pencarian grid terdiri dari mendefinisikan vektor batas bawah a = (a1, a2, …, am) dan vektor batas atas b = (b1, b2, …, bm) untuk setiap komponen . Pencarian grid melibatkan pengambilan n titik dengan jarak yang sama di setiap interval dari bentuk [ai, bi] termasuk ai dan bi. Ini menciptakan total nm titik kisi yang mungkin untuk diperiksa. Akhirnya, setelah setiap pasangan poin dihitung, maksimum dari nilai-nilai ini dipilih.

Masalah dengan jenis metode ini adalah bahwa jumlah evaluasi meningkat secara eksponensial dengan meningkatnya n dan m. Karena kita tidak dapat benar-benar mengurangi m, penurunan n adalah satu-satunya cara yang mungkin untuk memastikan bahwa metode berhenti dalam waktu yang wajar, tetapi ini mengurangi validitas solusi.

#### Cleaning Text
1. Fungsi cleaningText, menghapus mention, menghapus hashtag, menghapus RT, menghapus link, dan menghapus nomor, mengganti baris baru dengan spasi, menghapus semua tanda baca, dan menghapus berbagai macam karakter atau emoji teks dari kiri ke kanan. <br>

```python
def cleaningText(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) # remove mentions
    text = re.sub(r'#[A-Za-z0-9]+', '', text) # remove hashtag
    text = re.sub(r'RT[\s]', '', text) # remove RT
    text = re.sub(r"http\S+", '', text) # remove link
    text = re.sub(r'[0-9]+', '', text) # remove numbers

    text = text.replace('\n', ' ') # replace new line into space
    text = text.translate(str.maketrans('', '', string.punctuation)) # remove all punctuations
    text = text.strip(' ') # remove characters space from both left and right text
    return text
```

#### Case Folding Text

2. Fungsi caseFoldingText, mengubah semua huruf menjadi huruf kecil (lowercase) semua. Sementara itu, karakter lain yang bukan termasuk huruf dan angka, seperti tanda baca dan spasi dianggap sebagai delimiter. Delimiter ini bisa juga dihapus atau diabaikan dengan menggunakan perintah yang ada di Python.<br>

```python
def casefoldingText(text): # Converting all the characters in a text into lower case
    text = text.lower() 
    return text
```

#### Tokenizing Text
3. Fungsi tokenizingText, pemisahan string pada teks menjadi beberapa daftar token sehingga dapat digunakan untuk menganalisa teks <br>

```python
def tokenizingText(text): # Tokenizing or splitting a string, text into a list of tokens
    text = word_tokenize(text) 
    return text
```

#### Filtering Text
4. Fungsi filteringText, tahap mengambil kata-kata penting dari hasil token dengan menggunakan algoritma stoplist (membuang kata kurang penting) atau wordlist (menyimpan kata penting).


```python
def filteringText(text): # Remove stopwors in a text
    listStopwords = set(stopwords.words('indonesian'))
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    text = filtered 
    return text
```

#### Stemming Text
5. Fungsi stemmingText, proses menghilangkan infleksi kata ke bentuk dasarnya, namun bentuk dasar tersebut tidak berarti sama dengan akar kata (root word).

```python
def stemmingText(text): # Reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    text = [stemmer.stem(word) for word in text]
    return text
```

#### to Sentence
6. Fungsi toSentence, Ubah daftar kata menjadi kalimat


```python
def toSentence(list_words): # Convert list of words into sentence
    sentence = ' '.join(word for word in list_words)
    return sentence
```

Memanggil Semua fungsi Preprocessing untuk melakukan preprocessing data

```python
# Preprocessing tweets data

tweets['text_clean'] = tweets['tweet'].apply(cleaningText)
tweets['text_clean'] = tweets['text_clean'].apply(casefoldingText)
tweets.drop(['tweet'], axis = 1, inplace = True)

tweets['text_preprocessed'] = tweets['text_clean'].apply(tokenizingText)
tweets['text_preprocessed'] = tweets['text_preprocessed'].apply(filteringText)
tweets['text_preprocessed'] = tweets['text_preprocessed'].apply(stemmingText)

# drop duplicates/spams tweets
tweets.drop_duplicates(subset = 'text_clean', inplace = True)
```

## Bagging Classifier (Ensemble Learning)

### Bagging Classifier Tanpa Menggunakan Grid Search
Dalam menghitung hasil nilai Bagging classifier, tentu membutuhkan nilai sebuah parameter dari nilai bagging classifier agar bisa mendapatkan nilai akurasi yang baik. Dalam hal ini, bisa dilakukan dengan cara menentukan nilai paramaternya menggunakan grid seacrh. Namun disini saya bereksperimen menentukan nilai parameter dari nilai 1-500 menggunakan perulangan for sehingga mendapatkan nilai akurasi satu-satu dan nanti akan mencoba menggunakan grid search.

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
  
# load the data
#url = "https://raw.githubusercontent.com/MohSyarif190005/fileReadme/main/tweets_data_clean_label.csv"
#dataframe = pd.read_csv(url)
#arr = dataframe.values
X = X_train
Y = y_train
  
seed = 8
#kfold = model_selection.KFold(n_splits = 3,random_state = seed)
  
# initialize the base classifier
base_cls = DecisionTreeClassifier()

#Menyimpan Hasil Nilai Base Classifier
base_classifier=[]
hasilNilai_classifier=[]

for i in range (1, 100):
  # no. of base classifier
  num_trees = i
    
  # bagging classifier
  model = BaggingClassifier(base_estimator = base_cls,
                            n_estimators = num_trees,
                            random_state = seed)
    
  results = model_selection.cross_val_score(model, X, Y)

  #Nilai base classifier dan hasil nilai classifier disimpan dan akan ditampilkan di grafik
  base_classifier.append(i)
  hasilNilai_classifier.append(results.mean())

  print("accuracy :")
  print(results.mean())
```

Menampilkan data hasil perulangan For dan digrafikkan menggunakan Plot dari python

```python
import matplotlib.pyplot as plt

plt.plot(base_classifier, hasilNilai_classifier)
plt.title('Hasil Nilai Bagging Classifier')
plt.xlabel('Nilai Estimator')
plt.ylabel('Hasil Akurasi')
plt.grid(True)
plt.show()
```

Kelemahan menggunakan hanya for biasa ialah komputasinya yang lama dan keterbatasan menentukan batas agar komputasi yang dihasilkan tidak lama. Bisa dibilang ini hanya menggunakan perhitungan manual dan ini membutuhkan waktu komputasi yang lama.

### Bagging Classifier Menggunakan Grid Search


Nilai Base classifier jika menggunakan grid search dapat menentukan nilai estimator atau nilai yang terbaik sehingga nilai dari base classifier mendapatkan hasil akurasi yang terbaik. Dalam Grid Search, ada yang namanya param grid. Param Grid atau nilai parameter grid merupakan nilai dengan sejumlah nilai diskrit untuk masing-masing. <br><br>

Dapat digunakan untuk mengulangi kombinasi nilai parameter dengan iter fungsi bawaan Python. Urutan kombinasi parameter yang dihasilkan adalah deterministik.

```python
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import pandas as pd

param_grid = {
    'base_estimator__max_depth' : [1, 2, 3, 4, 5],
    'max_samples' : [0.05, 0.1, 0.2, 0.5]
}

X = X_train
Y = y_train

clf = GridSearchCV(BaggingClassifier(DecisionTreeClassifier(),
                                     n_estimators = 100, max_features = 0.5), param_grid)
 
results = model_selection.cross_val_score(clf, X, Y)
print("accuracy :")
print(results.mean())
```

### Error atau Evaluasi Hasil Bagging Clasifier
Hasil dari Bagging Classifier akan dihitung hasil error yang dihasilkan dengan menggunakan Grid Search.


```python
# Fit bc to the training set
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

# Evaluate acc_test
acc_test = accuracy_score(y_test, y_pred)
print('Hasil Error yang Dihasilkan: {:.2f}'.format(acc_test))
```

Random Forest adalah model ensemble kuat yang dibangun dengan sejumlah besar Pohon Keputusan. Ini mengatasi kekurangan dari pohon keputusan tunggal di samping beberapa keuntungan lainnya. Tapi itu tidak berarti bahwa itu selalu lebih baik daripada pohon keputusan. Ada beberapa contoh ketika pohon keputusan dapat berkinerja lebih baik daripada Random forest. <br>

Sebuah ensemble berfungsi sedemikian rupa sehingga setiap model dalam ensemble membantu mengimbangi kekurangan dari setiap model lainnya. Kita dapat mengatakan, jika hutan acak dibangun dengan 10 pohon keputusan, setiap pohon mungkin tidak berkinerja baik dengan data, tetapi pohon yang lebih kuat membantu mengisi celah untuk pohon yang lebih lemah. Inilah yang menjadikan ansambel model pembelajaran mesin yang kuat. <br> <br>

<img src='https://miro.medium.com/max/828/1*nwF3aAHaAQDiNMQMI_YhBg.webp'>

```python
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

# prediction on test set
y_pred=clf.predict(X_test)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Error Rate
rf_us = RandomForestClassifier()
rf_us.fit(X_train, y_train)

y_prediksi = rf_us.predict(X_test)
```

```python
#Import 
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
import warnings
warnings.filterwarnings('ignore')  # "error", "ignore", "always", "default", "module" or "once"
```

```python
print('Akurasi: ',accuracy_score(y_test, y_prediksi))
print('F1 Score: ',f1_score(y_test, y_prediksi, average='macro'))
print('Recall: ',recall_score(y_test, y_prediksi, average='macro'))
print('Presisi: ',precision_score(y_test, y_prediksi, average='macro'))
print('\n Hasil Klasifikasi:\n',classification_report(y_test, y_prediksi))
print('\n Confusion Matriks:\n',confusion_matrix(y_test, y_prediksi))
```

### Random Forest dengan Decision Tree
Decision tree merupakan sebuah metode dari machine learning dan termasuk kedalam supervised learning atau pembelajaran terawasi. Decision tree Itu membuat prediksi, seperti bagaimana, pikiran manusia akan membuat, dalam kehidupan nyata. Ini dapat dianggap sebagai serangkaian pernyataan if-then-else dan terus membuat keputusan atau prediksi di setiap titik, seiring pertumbuhannya. <br> 

Decision tree terlihat seperti diagram alur atau pohon terbalik. Tumbuh dari akar ke daun tetapi dengan cara terbalik. Kita dapat dengan mudah menginterpretasikan proses pengambilan keputusan / prediksi dari pohon keputusan, karena strukturnya. Berikut gambarannya: <br> <br>

<img src='https://miro.medium.com/max/750/1*vCOvJhXDBslmghC_vgRMIw.webp'> <br> <br>

Decision tree dapat berupa Pohon Klasifikasi atau Pohon Regresi, berdasarkan jenis variabel target. Kelas dalam kasus pohon klasifikasi didasarkan pada prediksi mayoritas di simpul daun. Dalam kasus regresi, nilai prediksi akhir didasarkan pada nilai rata-rata di simpul daun. <br> 

Ada berbagai algoritma Pohon Keputusan tergantung pada apakah itu split biner atau split multiway. Pemisahan Biner berarti simpul induk terbagi menjadi 2 simpul anak dan dalam pemisahan multiway, simpul induk terbagi menjadi lebih dari 2 simpul anak. Salah satu contoh dalam Decision tree yang berupa klasifikasi adalah CART (Classification and Regression Tree). <br>

CART atau Classification and Regression Tree berurusan dengan pohon split biner. Decision tree membagi data menjadi berbagai himpunan bagian dan kemudian membuat pemisahan berdasarkan atribut yang dipilih. Atribut ini dipilih berdasarkan kriteria homogenitas yang disebut Gini Index. <br>

Gini index pada dasarnya mengukur kemurnian (atau ketidakmurnian juga, kita dapat mengatakan) dari node setelah perpecahan terjadi. Artinya, ini adalah ukuran seberapa murni simpul anak dari simpul induk, setelah pemisahan. 

<img src='https://miro.medium.com/max/640/1*vEn0rakJ92Ppmhbd9d_KUQ.webp'> <br>

Pohon harus membuat pemisahan berdasarkan atribut yang mengarah ke homogenitas maksimum dari simpul anak. Rumus Gini Index adalah sebagai berikut :

<img src='https://miro.medium.com/max/828/1*Wnq44FUmqfzDqX6E6lPUMw.webp'> <br><br>

Dimana pi adalah probabilitas menemukan titik dengan label i, dan n adalah jumlah kelas/nilai resultan dari variabel target.

Dengan demikian pohon menghitung indeks Gini pada simpul induk. Kemudian indeks Gini tertimbang pada node anak dihitung untuk setiap atribut. Sebuah atribut dipilih sedemikian rupa sehingga kita mendapatkan homogenitas maksimum di node anak dan pemisahan dilakukan. Proses ini diulang sampai semua atribut habis. Pada akhirnya kita mendapatkan apa yang kita sebut sebagai pohon Keputusan yang tumbuh sepenuhnya. Dengan demikian atribut yang mengarah pada pemisahan awal dapat dianggap sebagai atribut yang paling penting untuk prediksi.

### Random Forest Menggunakan GridSearchCV
Kita melakukan penyetelan hyperparameter menggunakan GridSearchCV untuk model Random Forest Undersampling. Tuning hyperparameter serupa untuk Decision Tree Classifier. Kita hanya perlu mengganti Random Forest Classifier dengan Decision Tree Classifier, untuk model dasarnya.


```python
from sklearn.model_selection import GridSearchCV

hyper_params = {'max_depth': [3, 5, 10, 15, 20],
                'max_features': [3, 5, 7, 11, 15],
                'min_samples_leaf': [20, 50, 100, 200, 400],
                'n_estimators': [10, 25, 50, 80, 100]
                }
```

```python
#Grid search
model_cv = GridSearchCV(estimator=rf_us,
                        param_grid=hyper_params,
                        verbose=1,
                        cv=5,
                        n_jobs=1,
                        return_train_score=True)
```

```python
model_cv.fit(X_train, y_train)
```

```python
model_cv.best_score_
```

```python
model_cv.best_estimator_
```

```python
#Pengimplementasian best estimator hasil dari GridSearchCV
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(max_depth=3, max_features=3, min_samples_leaf=20,
                       n_estimators=10)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

# prediction on test set
y_pred=clf.predict(X_test)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
